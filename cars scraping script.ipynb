```python
#connecting to web driver
driver=webdriver.Chrome('chromedriver.exe')

url='https://www.google.com/search?rlz=1C1ZKTG_enIN965IN965&sxsrf=APq-WBsqAGuM3KYZNuSRSWRNsIKsVTd3hQ:1648017377571&q=cars24&spell=1&sa=X&ved=2ahUKEwjCpv6Sz9v2AhXZh1YBHX5jBb8QBSgAegQIAhAy&biw=1536&bih=722&dpr=1.25'
driver.get(url)
```

```python

# used car link click
usedcar_btn = driver.find_element_by_xpath ('/html/body/div[7]/div/div[10]/div[1]/div[1]/div[3]/div/div/div/div/div/div[1]/a/div[1]/span').click()
```

```python

# location click
location_btn = driver.find_element_by_xpath ('/html/body/div[1]/div[10]/div/div/div/div[2]/div[2]/button').click()
```

```python
# city selection click
city = driver.find_element_by_xpath ('/html/body/div[1]/div[10]/div/div/div/div[2]/ul/li[1]/span/img').click()
```

```python
# click for getting all cars detail
driver.execute_script("window.scrollBy(0,1000)") 
time.sleep(3)
all_ = driver.find_element_by_xpath ("//button[@class='btn _3BmFU AgO6q']/span").click()
driver.execute_script("window.scrollBy(0,1000)")
```

```python
#creating empty list

urls=[]
name = []
varient = []
price = []
emi = []
history=[]
last_serve=[]
purchase_on=[]
owner_no = []
fule = []
transmission = []
insuarance_upto = []
insua_typ=[]
Engine_type=[]
loc=[]
max_power =[]
estd_delivery=[]
travel_dist=[]

driver.execute_script("window.scrollBy(0,1000000)") 

time.sleep(35)
driver.execute_script("window.scrollBy(0,1000000)") 
#Fetching urls for each repository

url = driver.find_elements_by_xpath("//div[@class='col-4']/a")
for i in url:
    urls.append(i.get_attribute("href"))
    
```

```python
len(name)
```

```python
#Scraping data of car name 
nm = driver.find_elements_by_xpath("//h2[@class='_3FpCg']")
for i in nm:
    name.append(i.text)
    
#Scraping data of car var 
nm = driver.find_elements_by_xpath("//p[@class='cvakB']")
for i in nm:
    varient.append(i.text)
```

```python
   
#Scraping data from every repository page
for i in urls:
    driver.get(i)
    time.sleep(2)
    
    
    
    #Scraping data of car history
    try:
        his = driver.find_element_by_xpath("//li[@class='tHlIu'][1]")
        history.append(his.text.split('\n')[-1])
    except NoSuchElementException:
        history.append('None')
        
  
        
        
    #Scraping data of owners
    try:
        owner = driver.find_element_by_xpath("//li[@class='tHlIu'][2]")
        owner_no.append(owner.text.split('\n')[-1])
        
    except NoSuchElementException:
        owner_no.append('None')
        
        
    #fetching total traveled
    try:
        dist= driver.find_elements_by_xpath("//li[@class='tHlIu'][3]/strong")[0]
        travel_dist.append(dist.text.replace('km',''))
        
    except NoSuchElementException:
        travel_dist.append('None')
        
    #fetching fule type
    try:
        fl= driver.find_elements_by_xpath("//li[@class='tHlIu'][4]")[0]
        fule.append(fl.text.split('\n')[-1])
        
    except NoSuchElementException:
        fule.append('None')
        
    #fetching last served
    try:
        lst= driver.find_elements_by_xpath("//li[@class='tHlIu'][5]/strong")[0]
        last_serve.append(lst.text.replace('km','').split(' ')[0])
        
    except NoSuchElementException:
        last_serve.append('None')
        
    #fetching transmission
    try:
        trans= driver.find_elements_by_xpath("//li[@class='tHlIu'][6]")[0]
        transmission.append(trans.text.split('\n')[-1])
        
    except NoSuchElementException:
        transmission.append('None')
        
    #fetching insuarance
    try:
        ins= driver.find_elements_by_xpath("//li[@class='tHlIu'][8]")[0]
        insuarance_upto.append(ins.text.split('\n')[1].replace('Valid upto',''))
        
    except NoSuchElementException:
        insuarance_upto.append('None')
        
     #fetching insuarance type
    try:
        insu= driver.find_elements_by_xpath("//li[@class='tHlIu'][8]")[0]
        insua_typ.append(insu.text.split('\n')[-1].replace('Valid upto',''))
        
    except NoSuchElementException:
        insua_typ.append('None')
        
        
    #fetching total price
    try:
        prce= driver.find_elements_by_xpath("//strong[@class='_2yYvS Wqfj6']")[0]
        price.append(prce.text.split('\n')[-1].replace('₹',''))
        
    except NoSuchElementException:
        price.append('None')
        
    #fetching total emi available
    try:
        em= driver.find_elements_by_xpath("//strong[@class='_2yYvS']")[0]
        emi.append(em.text.split('\n')[-1].replace('₹','').replace('/month',''))
        
    except NoSuchElementException:
        emi.append('None')
        
    #fetching location
    try:
        lo= driver.find_elements_by_xpath("//p[@class='_2NHUv']/span")[0]
        loc.append(lo.text)
        
    except NoSuchElementException:
        loc.append('None')
        
     #fetching pickup
    try:
        pick= driver.find_elements_by_xpath("//p[@class='_3qCmx']/strong")[0]
        estd_delivery.append(pick.text)
        
    except NoSuchElementException:
        estd_delivery.append('None')
        
    
   
          
```
